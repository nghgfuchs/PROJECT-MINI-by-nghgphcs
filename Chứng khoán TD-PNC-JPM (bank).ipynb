### Phân tích chỉ số nhóm ngành của 3 ngân hàng

```python
import yfinance as yf
# import investpy
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
```

```python
start='2020-01-01'
end=dt.datetime.now().strftime('%Y-%m-%d')
TD = yf.download(tickers='TD',start=start,end=end) #Tải dữ liệu về df
TD #ngan hang TD Bank Mỹ
```

```python
PNC = yf.download(tickers='PNC',start='2020-01-01',end=dt.datetime.now().strftime('%Y-%m-%d')) #Tải dữ liệu về df
PNC #ngan hang PNC Bank Mỹ
```

```python
JPM = yf.download(tickers='JPM',start='2020-01-01',end=dt.datetime.now().strftime('%Y-%m-%d')) #Tải dữ liệu về df
JPM #ngan hang JPM Bank Mỹ
```

```python
# tạo 1 cái list bank
list_banks = ["TD","PNC","JPM"]
```

```python
# axis = 1 là nối theo cột
bank_stocks = pd.concat([TD,PNC,JPM],axis = 1,keys = list_banks) 
bank_stocks.head()
```

```python
bank_stocks.columns.names = ["Bank Name","Stock Info"]
bank_stocks.tail()
```

```python
# Giá trị đóng cửa lớn nhất của 3 ngân hàng từ năm start -> năm end
bank_stocks.xs(key='Close',axis = 1,level = 'Stock Info').max()
```

- Tạo một DataFrmae trống mới gọi là value_banks.DataFrame này chứa tỷ suất lợi nhuận cho từng cổ phiếu của các ngân hàng trên

```python
value_banks = pd.DataFrame()
```

Tạo 1 vòng lặp for để lặp qua với mỗi mã cổ phiếu ngân hàng

Tiếp theo, ta tạo 1 cột tên là "Mã cổ phiếu + value_banks" = giá tỷ suất lợi nhuận

Để tính tỷ suất lợi nhuận, ta dùng hàm pct_change trong pandas và áp dụng cho giá đóng cửa của từng mã cổ phiếu.
- Phương thức .pct_change() để tính toán thay đổi % theo thời gian. Sự thay đổi này được tính ( Giá trị row năm n - giá trị row năm n-1)/(giá trị row năm n-1). 

```python
for name in list_banks:
    value_banks[name+' value_bank'] = bank_stocks[name]['Close'].pct_change()*100
value_banks
```

```python
sns.pairplot(value_banks);
```

Dùng seaborn sử dụng biểu đồ pairplot(hay biểu đồ phân tán) để tìm cổ phiếu nổi bật dựa trên giá trị lợi nhuận bên trên

-> Giá trị % lợi nhuận càng cao 

```python
plt.figure(figsize = (15,5)) 
sns.distplot(value_banks.loc[start:end]['TD value_bank'],color = "blue",bins = 100);
#biểu đỒ phân phối lợi nhuận cổ phiếu của ngân hàng TD bank
```

```python
plt.figure(figsize = (15,5)) 
sns.distplot(value_banks.loc[start:end]['PNC value_bank'],color = "yellow",bins = 100);
#biểu đỒ phân phối lợi nhuận cổ phiếu của ngân hàng PNC bank
```

```python
plt.figure(figsize = (15,5)) 
sns.distplot(value_banks.loc[start:end]['JPM value_bank'],color = "red",bins = 100);
#biểu đỒ phân phối lợi nhuận cổ phiếu của ngân hàng JPM bank
```

### Đường trung bình động - MA
Đường MA là đường trung bình động, thể hiện biến động, chỉ báo xu hướng của giá cổ phiếu trong một khoảng thời gian. Mục đích của đường MA là theo giõi giá đang vận động theo xu hướng tăng hoặc giản hay không có xu hướng.

```python
# vẽ đường trung bình động và giá đóng của cổ phiếu TD = matplotlib
plt.figure(figsize=(12,6))
TD['Close'].loc[start:end].rolling(window=30).mean().plot(label="MA30") # hàm rolling là nó sẽ sử dụng dữ liệu của 30 ngày trước đó
TD['Close'].loc[start:end].plot(label="TD CLOSE")
plt.legend()
plt.show()
```
